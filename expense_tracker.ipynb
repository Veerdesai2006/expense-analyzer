```python
#listing expenses
expenses = [
    ("food", 150),
    ("travel", 100),
    ("food", 150),
    ("shopping", 500),
    ("travel", 100),
    ("food", 100)
]
```

```python
#Now we have to assign all the values in dictionary 
#We will take care that any values dosent go missing
expense_dict={}
for category, cost in expenses:
    if category in expense_dict:
        expense_dict[category].append(cost)
    else:
        expense_dict[category]=[cost]
        
total_expense=0
#Now we need to find the total expense and category expense

for category,cost in expense_dict.items():
    category_expense= sum(cost)
    total_expense+=category_expense
    print(category, " : " ,category_expense)
 

```
